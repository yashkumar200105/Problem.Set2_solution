# Question 3
Write a class named `Marsupial` that can be used as shown below:

```python
>>> m = Marsupial()
>>> m.put_in_pouch('doll')
>>> m.put_in_pouch('firetruck')
>>> m.put_in_pouch('kitten')
>>> m.pouch_contents()
['doll', 'firetruck', 'kitten']

Now write a class Kangaroo as a subclass of Marsupial that:
a) Extends the Marsupial __init__ constructor to take the coordinates x and y of the Kangaroo object.
b) Supports a jump method that takes values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the y-axis.
c) Overloads the __str__ operator to return the location as shown below:

>>> k = Kangaroo(0,0)
>>> print(k)
I am a Kangaroo located at coordinates (0,0)
>>> k.put_in_pouch('doll')
>>> k.put_in_pouch('firetruck')
>>> k.put_in_pouch('kitten')
>>> k.pouch_contents()
['doll', 'firetruck', 'kitten']
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> print(k)
I am a Kangaroo located at coordinates (3,0)


````python
class Marsupial:
    def __init__(self):
        self.pouch = []

    def put_in_pouch(self, item):
        self.pouch.append(item)

    def pouch_contents(self):
        return self.pouch

class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

### Testing

# Testing Marsupial
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())  # Expected output: ['doll', 'firetruck', 'kitten']

# Testing Kangaroo
k = Kangaroo(0, 0)
print(k)  # Expected output: I am a Kangaroo located at coordinates (0,0)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  # Expected output: ['doll', 'firetruck', 'kitten']
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  # Expected output: I am a Kangaroo located at coordinates (3,0)
````

# Question 4
Write a function `collatz()` that takes a positive integer `x` as input and prints the Collatz sequence starting at `x`. The Collatz sequence is obtained by repeatedly applying the following rule:

x = { x/2     if x is even
    { 3x + 1  if x is odd

The sequence stops when it reaches the number `1`. Implement `collatz()` recursively, without using any loops.

Example:
```python
>>> collatz(1)
1
>>> collatz(10)
10
5
16
8
4
2
1


````python
# Answer 4

def collatz(x):
    print(x)
    if x == 1:  # Base case: if x is 1, stop the recursion
        return
    elif x % 2 == 0:  # Even case: divide x by 2
        collatz(x // 2)
    else:  # Odd case: apply 3x + 1
        collatz(3 * x + 1)

# Testing

# Testing the Collatz sequence
collatz(1)   # Expected output: 1
collatz(10)  # Expected output: 10, 5, 16, 8, 4, 2, 1
````

# Question 5
Write a recursive method `binary()` that takes a non-negative integer `n` and prints the binary representation of integer `n`.

Example:
```python
>>> binary(0)
0
>>> binary(1)
1
>>> binary(3)
11
>>> binary(9)
1001

````python
# Answer 5
def binary(n):
    if n == 0:
        print(0, end='')
        return
    elif n == 1:
        print(1, end='')
        return
    else:
        binary(n // 2)  # Recursive case: divide n by 2
        print(n % 2, end='')  # Print the remainder (0 or 1)

# Testing binary representation
binary(0)  # Expected output: 0
print()    # For spacing
binary(1)  # Expected output: 1
print()    # For spacing
binary(3)  # Expected output: 11
print()    # For spacing
binary(9)  # Expected output: 1001
print()    # For spacing
````

# Question 6
Implement a class named `HeadingParser` to parse an HTML document, retrieving and printing all the headings (`<h1>`, `<h2>`, etc.) in the document. 

The class should:
- Be a subclass of `HTMLParser` (from the `html.parser` module).
- Print each heading on a new line in the order they appear in the document.
- Indent each heading based on its level (e.g., `<h1>` with 0 indentation, `<h2>` with 1 space, etc.).

Example usage:
```python
>>> infile = open('w3c.html')
>>> content = infile.read()
>>> infile.close()
>>> hp = HeadingParser()
>>> hp.feed(content)
W3C Mission
 Principles

````python
# Answer 6

from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.current_level = None  # To track the current heading level

    def handle_starttag(self, tag, attrs):
        if tag.startswith("h") and tag[1].isdigit():  # Check if it's an h1, h2, etc.
            self.current_level = int(tag[1])  # Set current heading level based on tag (1, 2, etc.)

    def handle_endtag(self, tag):
        if tag.startswith("h") and tag[1].isdigit():
            self.current_level = None  # Reset after heading ends

    def handle_data(self, data):
        if self.current_level is not None:  # If in a heading tag, print with indentation
            print(" " * (self.current_level - 1) + data.strip())

# Sample test with file 'w3c.html'
infile = open('w3c.html')
content = infile.read()
infile.close()

hp = HeadingParser()
hp.feed(content)
````

# Question 7
Implement a recursive function `webdir()` that:
- Takes a URL, `depth` (number of recursive link-following levels), and `indent` (initial indentation level).
- Visits and prints every reachable page up to the given depth from the starting URL.
- Indents each URL printed according to the depth of the link.

Example usage:
```python
>>> webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html


````python
# Answer 7
import requests
from html.parser import HTMLParser
from urllib.parse import urljoin

class LinkParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.links = []

    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for (attr, value) in attrs:
                if attr == 'href':
                    self.links.append(value)

    def get_links(self, url):
        try:
            response = requests.get(url)
            self.feed(response.text)
            return self.links
        except requests.exceptions.RequestException:
            return []

def webdir(url, depth, indent):
    if depth < 0:
        return

    # Print the current URL with indentation
    print(' ' * indent + url)

    # Parse links from the current page if depth allows
    parser = LinkParser()
    links = parser.get_links(url)

    # Recurse for each link found
    for link in links:
        absolute_url = urljoin(url, link)
        webdir(absolute_url, depth - 1, indent + 1)

# Test by calling webdir with sample URL and depth
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
````

Question 9
. Suppose list words is defined as follows:
>>> words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over',
'the', 'lazy', 'dog']
Write list comprehension expressions that use list words and generate the following
lists:
a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE',
'LAZY', 'DOG']
b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the',
'lazy', 'dog']
c) [3, 5, 5, 3, 5, 4, 3, 4, 3] (the list of lengths of words in list
words).
d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN',
'brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5],
['OVER', 'over', 4], ['THE', 'the', 3], ['LAZY', 'lazy',
4], ['DOG', 'dog', 3]] (the list containing a list for every word of list
words, where each list contains the word in uppercase and lowercase and the
length of the word.)
e) ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the',
'lazy', 'dog'] (the list of words in list words containing 4 or more
characters.)

````python
# Answer 9

# Given list
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# a) Words to uppercase
uppercase_words = [word.upper() for word in words]
# Result: ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE', 'LAZY', 'DOG']

# b) Words to lowercase
lowercase_words = [word.lower() for word in words]
# Result: ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# c) List of lengths of each word
lengths = [len(word) for word in words]
# Result: [3, 5, 5, 3, 5, 4, 3, 4, 3]

# d) List containing [uppercase, lowercase, length] for each word
word_details = [[word.upper(), word.lower(), len(word)] for word in words]
# Result: [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN', 'brown', 5], 
#          ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5], ['OVER', 'over', 4], 
#          ['THE', 'the', 3], ['LAZY', 'lazy', 4], ['DOG', 'dog', 3]]

# e) Words containing 4 or more characters
long_words = [word for word in words if len(word) >= 4]
# Result: ['quick', 'brown', 'jumps', 'over', 'lazy']

print("Words:",words)
print("Uppercase Words:",uppercase_words)
print("Lowercase Words:",lowercase_words)
print("Lengths:",lengths)
print("Word Details:",word_details)
print("Long Words: ",long_words)
````
